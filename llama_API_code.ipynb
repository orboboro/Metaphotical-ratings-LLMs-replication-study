```python
from ollama import chat
from ollama import ChatResponse

prompt = "Ti verrà presentata una lista di metafore. In una scala da 0 a 7, assegna un punteggio di familiarità e di efficacia per ciascuna metafora: " + "affetto è calore"

response: ChatResponse = chat(model='llama3', messages=[
  {
    'role': 'user',
    'content': prompt,
  },
])
print(response['message']['content'])
```
