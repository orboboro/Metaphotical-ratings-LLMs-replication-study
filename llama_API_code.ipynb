```python
from ollama import chat
from ollama import ChatResponse

prompt = "Agisci come se fossi un parlante nativo di italiano che deve esprimere giudizi quantitativi su espressioni metaforiche. Ti verrà presentata una lista di metafore. A ciascuna metafora assegna un valore di familiarità, cioè quanto ti è familiare quella metafora, e un valore di efficacia, cioè quanto secondo te è efficace quella metafora. Rispondi solo con numeri compresi tra 0 e 7, dove 0 vuol dire poco familiare o poco efficace e 7 vuol dire molto familiare o molto efficace." + "affetto è calore"

response: ChatResponse = chat(model='qwen:0.5b', messages=[
  {
    'role': 'user',
    'content': prompt,
  },
])
print(response['message']['content'])
```
